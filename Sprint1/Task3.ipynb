```
using System;

public class Person
{
    private string name;
    private int age;
    private string gender;

    public string Name
    {
        get { return name; }
        set { if (value == null)
                {
                    Console.WriteLine("Имя не может быть null!");
                }
                else
                {
                    name = value;
                }
            }
    }

    public int Age
    {
        get { return age; }
        set
        {
            if (value >= 0)
                age = value;
            else
                Console.WriteLine("Возраст не может быть отрицательным!");
        }
    }

    public string Gender
    {
        get { return gender; }
        set { if (value == null)
                {
                    Console.WriteLine("Пол не может быть null!");
                }
                else
                {
                    gender = value;
                }
            }
    }

    public Person() { }

    public Person(string name, int age, string gender)
    {
        Name = name;
        Age = age;
        Gender = gender;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Пол: {Gender}";
    }

    public virtual void SayHello()
    {
        Console.WriteLine($"Приветствую");
    }

    public void IntroduceTo(Person otherPerson)
    {
        Console.WriteLine($"{Name} говорит: Привет, {otherPerson.Name}!");
    }
}

public class Student : Person
{
    private string university;
    private string course;

    public string University
    {
        get { return university; }
        set {
                if (value == null)
                {
                    Console.WriteLine("Университет не может быть null!");
                }
                else
                {
                    university = value;
                }
            }
    }

    public string Course
    {
        get { return course; }
        set
        {
            if (string.IsNullOrEmpty(value))
                Console.WriteLine("Курс не может быть пустым!");
            else
                course = value;
        }
    }

    public Student() { }

    public Student(string name, int age, string gender, string university, string course)
        : base(name, age, gender)
    {
        University = university;
        Course = course;
    }

    public override void SayHello()
    {
        Console.WriteLine($"Привет, меня зовут {Name}, я учусь на {Course} в {University}");
    }
}

public class Employee : Person
{
    private string company;
    private double salary;

    public string Company
{
    get { return company; }
    set
    {
        if (value == null)
        {
            Console.WriteLine("Компания не может быть null!");
        }
        else
        {
            company = value;
        }
    }
}


    public double Salary
    {
        get { return salary; }
        set
        {
            if (value >= 0)
                salary = value;
            else
                Console.WriteLine("Зарплата не может быть отрицательной!");
        }
    }

    public Employee() { }

    public Employee(string name, int age, string gender, string company, double salary)
        : base(name, age, gender)
    {
        Company = company;
        Salary = salary;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}, компания: {Company}, зарплата: {Salary:C}";
    }

    public void WorkWith(Employee colleague)
    {
        Console.WriteLine($"{Name} и {colleague.Name} работают вместе в компании {Company}.");
    }
}

public class Teacher : Person
{
    private string subject;
    private int experience;

    public string Subject
    {
        get { return subject; }
        set { if (value == null)
                {
                    Console.WriteLine("Предмет не может быть null!");
                }
                else
                {
                    subject = value;
                } 
            }
    }

    public int Experience
    {
        get { return experience; }
        set
        {
            if (value >= 0)
                experience = value;
            else
                Console.WriteLine("Опыт не может быть отрицательным!");
        }
    }

    public Teacher() { }

    public Teacher(string name, int age, string gender, string subject, int experience)
        : base(name, age, gender)
    {
        Subject = subject;
        Experience = experience;
    }

    public override void SayHello()
    {
        Console.WriteLine($"Привет, меня зовут {Name}, я преподаю дисциплину {Subject} уже {Experience} лет");
    }

    public void TeachStudent(Student student)
    {
        Console.WriteLine($"{Name} преподает {Subject} студенту {student.Name}, который учится на {student.Course}.");
    }
}
```

```

Person person = new Person("Игорь Иванов", 35, "Мужчина");
Student student = new Student("Алиса Жирикова", 22, "Женщина", "Тюменский индустриальный университет", "Программирование");
Employee employee = new Employee("Арсений Васенин", 25, "Мужчина", "ООО \"СуперПупер\"", 50000);
Teacher teacher = new Teacher("Ольга Рукова", 40, "Женщина", "Математика", 20);

person.IntroduceTo(student);
student.SayHello();
employee.WorkWith(employee); 
teacher.TeachStudent(student);
```
